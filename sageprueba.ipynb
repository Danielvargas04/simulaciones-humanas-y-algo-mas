```python
def funcion_exponencial(x):
    return exp(x)
def funcion_exponencial2(x):
    return exp(x/2)

```

```python
grafico = plot(funcion_exponencial, (-5, 5), color = 'orange', legend_label='funcion exponencial',
               axes_labels=('x','y'), title='graficos')
grafico += plot(funcion_exponencial2, (-5, 5), color = 'blue', legend_label='funcion exponencial',
               axes_labels=('x','y'), title='graficos')
grafico.show()
```
